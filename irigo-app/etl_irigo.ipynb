```python
# pour les date
import datetime as dt

# Pour manipuler efficacement des tables de données dans Python
import pandas as pd

# Pour faire des requêtes GET et POST
import requests

# sql
from sqlalchemy import create_engine
```

```python
r = requests.get("https://data.angers.fr/api/records/1.0/search/",
                 params={
                     'dataset': 'bus-tram-position-tr',
                     'rows': 10,
                     'facet':'novh',
                     'facet':'mnemoligne',
                     'facet':'nomligne',
                     'facet':'dest'
                 })
```

```python
type(r)
```

```python
r.status_code
```

```python
# r.json()
```

```python
dd = r.json(parse_int=str)['records']
```

```python
# dd[0]
```

```python
# récupération des données du json

#bus
idBus = [elem['fields']['idvh'] for elem in dd]
type = [elem['fields']['type'] for elem in dd]
etat = [elem['fields']['etat'] for elem in dd]

#arret
idArret = [elem['fields']['idarret'] for elem in dd]
nomarret = [elem['fields']['nomarret'] for elem in dd]

#ligne
idLigne  = [elem['fields']['idligne'] for elem in dd]
nomLigne = [elem['fields']['nomligne'] for elem in dd]
numLigne = [elem['fields']['mnemoligne'] for elem in dd]

#trajet
lats = [elem['fields']['coordonnees'][0] for elem in dd]
lons = [elem['fields']['coordonnees'][1] for elem in dd]
dest = [elem['fields']['dest'] for elem in dd]

#etape
harret = [elem['fields']['harret'] for elem in dd]
ecart = [elem['fields']['ecart'] for elem in dd]

df = pd.DataFrame({
    'idBus': idBus,
    'type': type,
    'etat': etat,
    'idArret': idArret,
    'nomarret': nomarret,
    'idLigne': idLigne,
    'nomLigne': nomLigne,
    'numLigne': numLigne,
    'lats': lats,
    'lons': lons,
    'dest': dest,
    'harret': harret,
    'ecart': ecart
})

# df
```

```python
# df.info()
```

```python
# ouverture de la connection vers la bdd
engine = create_engine("sqlite:///data.sqlite")
connection = engine.connect()
```

```python
# bus
dft = df[['idBus', 'type', 'etat']]
print(dft)
```

```python
assert not df['idBus'].isnull().any()
```

```python
# export dans SQL
dft.to_sql('bus', connection, if_exists='replace', index=False)
```

```python
# arret
dft = df[['idArret', 'nomarret']]
print(dft)
```

```python
# export dans SQL
dft.to_sql('arret', connection, if_exists='replace', index=False)
```

```python
# ligne
dft = df[['idLigne', 'nomLigne', 'numLigne']]
print(dft)
```

```python
# export dans SQL
dft.to_sql('ligne', connection, if_exists='replace', index=False)
```

```python
# trajet
dft = df[['idBus', 'idLigne', 'lats', 'lons', 'dest']]
print(dft)
```

```python
# export dans SQL
dft.to_sql('trajet', connection, if_exists='replace', index=False)
```

```python
# conversion harret
df['harret'] = pd.to_datetime(df['harret'], format='%Y-%m-%d %H:%M:%S.%f')
print(df[['harret']])
```

```python
# convertir ecart en int
df['ecart'] = df['ecart'].astype(int)
```
